```python
%matplotlib inline
import netCDF4
from netCDF4 import Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from windrose import WindroseAxes
import matplotlib.cm as cm
```

```python
ff = Dataset('/home/bikas/Desktop/ERA_data/ERA_analysis/temp_water_vap.nc')

def era_read(file):
    Lon = file.variables['longitude'][:]
    time = file.variables['time']
    Lat= file.variables['latitude'][:]
    dates = netCDF4.num2date(time[:], time.units, time.calendar)
    
    u10 = np.average(np.average(file.variables['u10']  [:,30:31,39:40], axis = 1),axis = 1)
    v10 = np.average(np.average(file.variables['v10']  [:,30:31,39:40], axis = 1),axis = 1)
    speed = np.sqrt(u10**2+v10**2)
    
    
    def toDirection(u_component, v_component):
        degrees = np.degrees(np.arctan2(-u_component, -v_component))
        degrees = np.where(degrees < 0, degrees + 360, degrees)
        return degrees
    
    
    direction = toDirection(u10,v10)
    df = pd.DataFrame({'u10':u10,'v10':v10,'speed':speed,'dir':direction})
    
    df = df.set_index(dates,drop=True, append=False, inplace=False, verify_integrity=False)
    df = df['2000-2-25':'2015-12-31'].resample('d').mean()
    return df

df_a = era_read(ff)
```

```python
def toDirection(u_component, v_component):
    """Convert components to wind direction.
    Args:
        u_component: u wind component float
        v_component: v wind component float
    Returns:
        wind direction in degrees [0.0;360.0[ rounded to nearest 0.5
    """
    degrees = np.degrees(np.arctan2(-u_component, -v_component))
    degrees = np.where(degrees < 0, degrees + 360, degrees)
    #degrees[degrees < 0] += 360
    return degrees

def toSpeed(u_component, v_component):
    """Returns the norm of the vectors, i.d., the speed.
    """
    return np.hypot(u_component,v_component)
```

```python
#Create wind speed and direction variables
ws = df_a['speed']
wd = df_a['dir']

#A quick way to create new windrose axes...
def new_axes():
    fig = plt.figure(figsize=(8, 8), dpi=80, facecolor='w', edgecolor='w')
    rect = [0.1, 0.1, 0.8, 0.8]
    ax = WindroseAxes(fig, rect, axisbg='w')
    fig.add_axes(ax)
    return ax

#...and adjust the legend box
def set_legend(ax):
    l = ax.legend(borderaxespad=-0.10)
    plt.setp(l.get_texts(), fontsize=8)

#A stacked histogram with normed (displayed in percent) results :

ax = new_axes()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
set_legend(ax)
```

```python
aod_dry = pd.read_table('aod_dry.csv',sep = ',',index_col = [0]).astype('float64')
aod_dry.head(3)
```

```python
#Remove outliers from AOD datasets
myList = aod_dry.values[:,0]

mean_duration  = np.nanmean(myList)
std_dev_one_test = np.nanstd(myList)     

def drop_outliers(x):
    if abs(x - mean_duration) <= std_dev_one_test:
        return x
myList = list(filter(drop_outliers, myList))
```

```python
plt.boxplot(myList)
plt.show()
```
